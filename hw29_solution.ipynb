```python
import numpy as np 
import pandas as pd
from IPython.display import display

print('========== read mpg miles per gallon data on cars =======================')
mpg_df = pd.read_csv(r"sport_athletes.csv")

mpg_df.groupby('Sport Type').max(numeric_only=True).rename(columns={'Weight(kg)':'max_weight'})
mpg_df.groupby('Sport Type').agg(max_weight=('Weight(kg)', 'max'))

mpg_df.groupby('Sport Type')[['Weight(kg)']].min().sort_values(by='Weight(kg)')

mpg_df.groupby('Sport Type')['Height(cm)'].nlargest(2)

mpg_df.groupby(['Country', 'Sport Type']).max()

#mpg_df.groupby(['Country', 'Sport Type']).max().loc[('USA', 'Basketball')]

#mpg_df.groupby(['Country', 'Sport Type']).max().loc[('USA', 'Basketball')]

#mpg_df.groupby(['Country', 'Sport Type']).max().loc[('Brazil','Soccer'):('Netherlands', 'Cycling')]
mpg_df.groupby(['Country', 'Sport Type']).max().xs(level='Sport Type', key='Soccer')

mpg_df[mpg_df['Sport Type'].isin(['Soccer', 'Tennis'])].groupby(['Country', 'Sport Type']).max()

mpg_df.groupby(['Country', 'Sport Type']).max().swaplevel().sort_values(by='Sport Type')
```
