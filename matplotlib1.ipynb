```python
import numpy as np 
import pandas as pd
from IPython.display import display

import matplotlib.pyplot as plt
x = np.arange(0, 10)
y = 2 * x
display(x)
display(y)

plt.plot(x, y)
plt.title("Hours vs Production")
plt.xlabel('Hours')
plt.ylabel('Production')
plt.legend(['Production'])
plt.xlim(0, 6)
plt.ylim(0, 12.5)

plt.savefig('line_chart1.jpg')

plt.show()
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(6.4, 4.8))  # 640 ==> 6.4, 480 => 4.8

#            left=0 bottom=0 width=% height=%
axes1 = fig.add_axes([0, 0, 1, 1])
x = np.arange(0, 10)
y = 2 * x
axes1.set_xlim(0, 6)
axes1.set_ylim(0, 12.5)
axes1.set_xlabel('Hours')
axes1.set_ylabel('Production')
axes1.plot(x, y)

axes2 = fig.add_axes([0.2, 0.5, 0.25, 0.3])
a = np.arange(0, 10)
b = a ** 4
axes2.set_title('Zoomed In')
axes2.plot(a, b)

axes3 = fig.add_axes([1.2, 0, 1, 1])
a = np.arange(0, 10)
b = a ** 4
axes3.plot(a, b)

plt.show()
```
